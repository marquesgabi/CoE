```python
#name_video = 'VideoN3.avi'
#name_video = 'VideoN2.avi'
name_video = 'strawberry.mp4'
cap = cv2.VideoCapture(name_video)


width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))


# count the number of frames
frames = cap.get(cv2.CAP_PROP_FRAME_COUNT)
fps = cap.get(cv2.CAP_PROP_FPS)
 
# calculate duration of the video
seconds = round(frames / fps)
video_time = datetime.timedelta(seconds=seconds)
print(f"duration in seconds: {seconds}")
print(f"video time: {video_time}")
print('frame per seconds:',fps)
print('number of frames:',frames)
```

```python
cont = 0 
k = 0
filtro =  400
size_x, size_y = 600, 300
if cap.isOpened() == False:
  print('video not found')

while cap.isOpened():
  cont = cont + 1
  ret,frame =cap.read()

  if ret == True:
    if(cont % filtro == 0):
      k += 1
      img = cv2.resize(frame, (size_x, size_y))
      cv2_imshow(img)
      cv2.imwrite("frame%d.jpg" % k, frame)

    if cv2.waitKey(10) & 0xFF == ord('q'):
      break
  else:
    break


cap.release()
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(name_video)

cont = 0 
k = 0
filtro =  400
size_x, size_y = 600, 300
if cap.isOpened() == False:
  print('video not found')

while cap.isOpened():
  cont = cont + 1
  ret,frame =cap.read()

  if ret == True:
    if(cont % filtro == 0):
      k += 1
      img = cv2.resize(frame, (size_x, size_y))
      gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
      #blur it
      blurred_image = cv2.GaussianBlur(gray, (7,7), 0)
      canny = cv2.Canny(blurred_image, 10, 30)
      cv2_imshow(canny)
      #cv2.imwrite("frame%d.jpg" % k, frame)

    if cv2.waitKey(10) & 0xFF == ord('q'):
      break
  else:
    break


cap.release()
cv2.destroyAllWindows()
```
